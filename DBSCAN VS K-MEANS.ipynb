```python
group = []
for k in range(2,10):
    model = KMeans(n_clusters=k)
    model.fit(eye)
    group.append(model.inertia_)

plt.plot(range(2,10),group,'o--')
plt.xlabel('K-values')
plt.ylabel('SSD')
plt.title('SSD vs K');
```

### USING DBSCAN on both tables

```python
out_per = []

for eps in np.linspace(0.001,3,50):
    
    model = DBSCAN(eps=eps)
    model.fit(eye[['feature1','feature2']])
    
    out = np.sum(model.labels_ == -1)
    
    per = 100 * out/len(model.labels_)
    
    out_per.append(per)

sns.lineplot(x= np.linspace(0.001,3,50),y=out_per)
plt.ylabel("Percentage of Points Classified as Outliers")
plt.xlabel("Epsilon Value")
plt.ylim(0,5)
plt.xlim(0,2)
plt.hlines(y=1,xmin=0,xmax=2,colors='red',ls='--');
```

LETS ASSUME OUR DATA HAS 1 percent points as outlers.SO epsilon value here is 0.85

```python
model = DBSCAN(eps=0.85)

eye['db_label'] = model.fit_predict(eye[['feature1','feature2']])

sns.scatterplot(data=eye,x='feature1',y='feature2',hue='db_label',palette='Set1')
plt.legend(loc=(1.05,0.5));
```

```python
out_per = []

for eps in np.linspace(0.001,3,50):
    
    model = DBSCAN(eps=eps)
    model.fit(ears[['feature1','feature2']])
    
    out = np.sum(model.labels_ == -1)
    
    per = 100 * out/len(model.labels_)
    
    out_per.append(per)

sns.lineplot(x= np.linspace(0.001,3,50),y=out_per)
plt.ylabel("Percentage of Points Classified as Outliers")
plt.xlabel("Epsilon Value")
plt.ylim(0,5)
plt.xlim(1.5,4)
plt.hlines(y=1,xmin=0,xmax=5,colors='red',ls='--');
```
